### Tratamiento de release_data

En la anterior se elimino de **release_date** porque los datos se sesgaban, aparte que lla informacion es suficiente para poder realizar el analisis correspondiente, pero en caso de no eliminarse se usara el siguiente bloque de codigo para rellenar los nulos con la mediana, pero se pudo evidenciar que los datos se sesgaban

## Transformaciones en User Reviews

```python
#leer user_reviews
df_user_reviews = pd.read_csv("Dataset/user_reviews.csv", encoding='utf-8')
```

### Eliminamos duplicados

```python
df_user_reviews = df_user_reviews.drop_duplicates()
df_user_reviews
```

### Columna review convertirlo a sentiment_analysis

usaremos la libreria nltk

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
df_user_reviews['review'] = df_user_reviews['review'].astype(str)
```

creamos la funcion para convertirlo a Negativo = 0, Neutral = 1 y Positivo =2

```python
def convertir_sentimiento(valor):
    if pd.isnull(valor):
        return 0
    elif valor < -0.2:
        return 0
    elif valor >= 0.2:
        return 2
    elif -0.2 <= valor < 0.2:
        return 1
    else:
        return None
```

Creamos la columna sentiment_analysis en donde sera el sentimiento de la review

```python
sia = SentimentIntensityAnalyzer()
df_user_reviews["sentiment_analysis"] = df_user_reviews["review"].apply(lambda i: convertir_sentimiento(sia.polarity_scores(i)['compound']))
df_user_reviews.head()
```

Ahora podemos ver los datos cambiados que nos serviran para el analisis

```python
df_user_reviews.info()
```

### Columna Posted

En esta columna estraremos el año y de aquellos que no tiene año usaremos la media para rellenar los datos

```python
# Extraer el año de la columna 'posted' y crear una nueva columna 'anio'
df_user_reviews['anio'] = df_user_reviews['posted'].str.extract(r'(\d{4})', expand=False)

# Convertir la columna 'anio' a tipo numérico
df_user_reviews['anio'] = pd.to_numeric(df_user_reviews['anio'], errors='coerce')

# Calcular la mediana de los años
mediana_anio = df_user_reviews['anio'].median()

# Llenar los valores nulos en 'anio' con la mediana calculada
df_user_reviews['anio'].fillna(mediana_anio, inplace=True)

# Convertir la columna 'anio' a tipo entero
df_user_reviews['anio'] = df_user_reviews['anio'].astype(int)

# Eliminar la columna 'posted'
df_user_reviews.drop('posted', axis=1, inplace=True)
df_user_reviews
```

## Transformaciones para ciertos endpoints

A la hora de deployar la api se tiene problemas con la memoria por tanto se decidio generar tablas pivot que agrupen la informacion

### PlaytimeGenre

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
#leemos el archivo
df_steam_games1 = pd.read_json('Dataset/output_steam_games.json', lines=True)
# Busca las filas con todas las columnas nulas y las elimina
df_steam_games1 = df_steam_games1.dropna(how='all')
# Reseteamos el index
df_steam_games1 = df_steam_games1.reset_index(drop=True)
#eliminamos columnas
df_steam_games1 = df_steam_games1.drop(["url", "price", "reviews_url",'specs', 'early_access',"tags","publisher", "app_name","title","developer" ], axis=1)
df_steam_games1 = df_steam_games1.rename(columns={'id': 'item_id'})
df_steam_games1 = df_steam_games1.drop([74,14573,30961]).reset_index(drop=True)
df_steam_games1 = df_steam_games1.drop([2579]).reset_index(drop=True)
# Comprobar y manejar los valores nulos en 'release_date'
df_steam_games1['anio'] = pd.to_datetime(df_steam_games1['release_date'], errors='coerce')

# Eliminar filas con valores nulos en 'release_date'
df_steam_games1 = df_steam_games1.dropna(subset=['anio'])

# Extraer el año de 'release_date' y convertir a tipo entero
df_steam_games1['anio'] = df_steam_games1['anio'].dt.year.astype(int)

# Convertir 'item_id' a tipo entero
df_steam_games1['item_id'] = df_steam_games1['item_id'].astype(int)

# Eliminar la columna 'release_date'
df_steam_games1 = df_steam_games1.drop('release_date', axis=1)

# Cambiar el nombre de 'anio' a 'release_date'
df_steam_games1 = df_steam_games1.rename(columns={'anio': 'release_date'})
df_steam_games1['genres'].fillna("sin dato", inplace=True)
df_steam_games1 = df_steam_games1.explode('genres', ignore_index=True)
# merge entre los dataframe user_items y steam_games
df_combinado =pd.merge(df_steam_games1 ,df_user_items, on='item_id', how='inner')
# Crear la tabla pivot
tabla_pivot_steam_user_items = df_combinado.pivot_table(
    values='playtime_forever',
    index='release_date',
    columns='genres',
    aggfunc=np.sum,
    fill_value=0
).round(2)

tabla_pivot_steam_user_items
```

```python
# Exportamos la tabla pivot
tabla_pivot_steam_user_items.to_parquet("Dataset/tabla_pivot_steam_user_items.parquet")
```

### UserForGenre

```python
# tabla pivote para mezclar user_id y genero y de valor se suman los playtime en cada iteraccion
df_genero_user_id=df_combinado.pivot_table(
    values='playtime_forever',
    index='user_id',
    columns='genres',
    aggfunc=np.sum,
    fill_value=0
).round(2)
df_genero_user_id
```

```python
# tabla pivote para mezclar user_id y genero y de valor se suman los playtime en cada iteraccion
df_user_id_release_date=df_combinado.pivot_table(
    values='playtime_forever',
    index='user_id',
    columns='release_date',
    aggfunc=np.sum,
    fill_value=0
).round(2)
df_user_id_release_date
```
